```python
#Display rows index 15-27
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cars.csv')

df.iloc[15:28]
```

```python
#Using describe method
df.describe()
```

```python
#Display number of rows and columns
df.shape
```

```python
#Display average price between fueltypes using a barplot
df.groupby('fueltype')['price'].mean().plot(kind= 'bar')
plt.title('Price vs Fuel Type')
plt.ylabel('Average Price')
plt.xlabel('Fuel Type')
```

```python
#Display average price between door numbers using a barplot
print(df.groupby('doornumber')['price'].mean())
```

**Based on the average price calculated for features in (4) and (6), what is your conclusion?** <br>

Based on the calculated data, we can see that the average price features in doors numbered 4 is greater than the
average price of features in doors numbered 2. Which means, on average, that doors numbered 4 will be generally more expensive.

```python
#Creating a scatter plot from the variables: horse power, highway mpg, and drive wheel
sns.scatterplot(data= df, x= 'horsepower', y= 'highwaympg', hue= 'drivewheel')
plt.show()
```
